```python
#Decoding Transaction Dynamics on PhonePe
#1. Total transaction amount and count by state

sql1 = """ SELECT state, 
       SUM(transacion_count) AS total_transactions, 
       SUM(transacion_amount) AS total_value
FROM agg_trans
GROUP BY state
ORDER BY total_value DESC;
"""
mycursor.execute(sql1)

data=mycursor.fetchall()
df = pd.DataFrame(data, columns=['state',"Total_Transactions","Total_Transaction_Amount"])
df
```

```python
#2. Quarterly transaction trends by payment category

sql2 = """ SELECT year, quater, transacion_type, 
       SUM(transacion_count) AS transaction_count, 
       SUM(transacion_amount) AS transaction_amount
FROM agg_trans
GROUP BY year, quater, transacion_type
ORDER BY year, quater;

"""
mycursor.execute(sql2)

data=mycursor.fetchall()
df = pd.DataFrame(data, columns=['Year','Quarter','Transaction_Type',"Total_Transactions","Total_Transaction_Amount"])
df
```

```python
#3. Top 5 states with highest average transaction value

sql3 = """ SELECT state, 
       ROUND(SUM(transacion_amount) * 1.0 / SUM(transacion_count), 2) AS avg_transaction_value
FROM agg_trans
GROUP BY state
ORDER BY avg_transaction_value DESC
LIMIT 5;
"""
mycursor.execute(sql3)

data=mycursor.fetchall()
df = pd.DataFrame(data, columns=['State','Avg_Transaction_Value'])
df

```

```python
#4. Quarter-over-quarter transaction growth by state

sql4 = """ WITH trans_by_quarter AS (
  SELECT state, year, quater, SUM(transacion_amount) AS total_amount
  FROM agg_trans
  GROUP BY state, year, quater
),
growth_calc AS (
  SELECT state, year, quater, total_amount,
         LAG(total_amount) OVER (PARTITION BY state ORDER BY year, quater) AS prev_amount
  FROM trans_by_quarter
)
SELECT state, year, quater, 
       prev_amount, total_amount, 
       ROUND((total_amount - prev_amount) * 100.0 / NULLIF(prev_amount, 0), 2) AS growth_percentage
FROM growth_calc
WHERE prev_amount IS NOT NULL
ORDER BY growth_percentage DESC;

"""
mycursor.execute(sql4)

data=mycursor.fetchall()
df = pd.DataFrame(data, columns=['State','Year','Quarter','Prev_amount', 'Total_amount', 'Growth_Percentage'])
df
```

```python
#5. States showing decline in transaction value
sql5 = """ WITH state_growth AS (
  SELECT state, year, quater, SUM(transacion_amount) AS total_amount
  FROM agg_trans
  GROUP BY state, year, quater
),
decline_calc AS (
  SELECT state, year, quater, total_amount,
         LAG(total_amount) OVER (PARTITION BY state ORDER BY year, quater) AS prev_amount
  FROM state_growth
)
SELECT state, year, quater, prev_amount, total_amount,
       ROUND((total_amount - prev_amount) * 100.0 / NULLIF(prev_amount, 0), 2) AS growth_pct
FROM decline_calc
WHERE total_amount < prev_amount
ORDER BY growth_pct;
"""

mycursor.execute(sql5)

data=mycursor.fetchall()
df = pd.DataFrame(data, columns=['State','Year','Quarter','Prev_amount', 'Total_amount', 'Growth_Percentage'])
df
```
